```python
import sys

from matplotlib import pyplot as plt
import numpy as np
from scipy.optimize import fsolve

sys.path.append("../../")
from analytical_beam_modes import cantilever_axial_modes, cantilever_bending_modes, cantilever_twisting_modes

sys.path.append("../")
from ehermes_beam import bending_beam as beam

%matplotlib inline
```

## Cantilever Bending in ZY plane

```python
N = 5
bendingz_modes = cantilever_bending_modes(beam, N)
bendingz_freqs = map(lambda x: x[0], bendingz_modes)

plt.plot(range(N), bendingz_freqs, 'bo-')
```

```python
x = np.linspace(0, 1, 100)
for freq, mode in bendingz_modes:
    plt.plot(mode / mode[-1])
plt.legend(["f = %.2f Hz" % x for x in bendingz_freqs], loc='center right', bbox_to_anchor=(1.45, 0.5))
```

## First 20 natural frequencies

```python
N = 20

bendingz_modes = cantilever_bending_modes(beam, N)
bendingx_modes = cantilever_bending_modes(beam, N, vertical='x')
axial_modes = cantilever_axial_modes(beam, N)
twisting_modes = cantilever_twisting_modes(beam, N)

all_modes = map(lambda x: (x[0], "bending_z"), bendingz_modes)
all_modes.extend(map(lambda x: (x[0], "bending_x"), bendingx_modes))
all_modes.extend(map(lambda x: (x[0], "axial"), axial_modes))
all_modes.extend(map(lambda x: (x[0], "twisting"), twisting_modes))
all_modes.sort(key=lambda x: x[0])

for mode, kind in all_modes[:N]:
    print "f = %.2f Hz (%s)" % (mode, kind)
```
